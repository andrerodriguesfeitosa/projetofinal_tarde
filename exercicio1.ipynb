- Select JOIN - Tabela aluno com tabela avaliação.

```python
SELECT
	a.id,
    a.ra,
    av.nota1,
    av.nota2,
    av.nota3,
    av.nota4

FROM
	aluno AS a
JOIN
	avaliacao AS av
ON
	av.fk_aluno_id = a.id;
```
